```python
import os
from gensim.models import FastText
```

```python
f=open("vocab.txt")
sentences=[line.split() for line in f]
```

```python
model=FastText(
  sentences=sentences,
  vector_size=250,
  window=10,
  min_count=2,
  sg=1
)
model.build_vocab(sentences)
model.train(
  sentences,
  total_examples=len(sentences),
  epochs=10
)
```

```python
print(model.wv.most_similar("wolf"))
```
